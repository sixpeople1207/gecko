```python
import requests 
from datetime import datetime
from datetime import date
import time
import sqlite3


path = './CoinData_Final_NewNew.db'

def get_histroy(name, fromDate, toDate):
    url = f'https://api.coingecko.com/api/v3/coins/{name}/market_chart/range?vs_currency=usd&from={fromDate}&to={toDate}&precision=0.00'
    print(url)
    response = requests.get(url).json();
    return response

def get_list():
    url = 'https://api.coingecko.com/api/v3/coins/markets/?vs_currency=usd&order=market_cap_desc&per_page=300&page=1&sparkline=false'
    response = requests.get(url).json()
    return response

def DBInIt(path):
    con = sqlite3.connect(path)
    cur = con.cursor()
    cur.execute("CREATE TABLE CoinList(Rank text,Name text,ath text);")
    cur.execute("CREATE TABLE CoinHistory(Name text, Date DateTime,Price float);")

def DBDRopTable(path, tableName):
    con = sqlite3.connect(path)
    cur = con.cursor()
    cur.execute(f"drop table {tableName}")

#DBDRopTable(path, 'CoinList')
#DBDRopTable(path, 'CoinHistory')
DBInIt(path)
```

```python
def DBInsert_CoinList(path, list):
    con = sqlite3.connect(path)
    cur = con.cursor()
    for li in list:
        #print("INSERT INTO CoinList Values('"+li[0]['market_cap_rank']+"','"+li[0]['id']+"','"+li[0]['ath']+"');")
        cur.execute(f"INSERT INTO CoinList Values('{li['market_cap_rank']}','{li['id']}','{li['ath']}');")
    con.commit()
    con.close()


DBInsert_CoinList(path, nameList)
```

```python
def DBInsert_Prices(path, nameList):
    con = sqlite3.connect(path)
    cur = con.cursor()
    for name in nameList:
        history=get_histroy(name['id'], str(fromDate), str(toDate))
        time.sleep(10)
        for i in range(len(history['prices'])):
            date = datetime.fromtimestamp(history['prices'][i][0]/1000)
            cur.execute(
                f"INSERT INTO CoinHistory Values('{name['id']}', '{date}','{history['prices'][i][1]}');")
        con.commit()
        print(name)
    con.close()

DBInsert_Prices(path, nameList)
```
